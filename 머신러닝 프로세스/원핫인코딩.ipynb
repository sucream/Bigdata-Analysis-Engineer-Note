# 범주변수의 변환
- 원핫인코딩

```python
import pandas as pd
```

```python
data = pd.read_csv('../data/vote.csv')
data.head()
```

## 성별과 지역
- 범주형 데이터가 수치형 데이터로 느껴질 수 있음
- 원핫인코딩을 통해 dummy variable을 만들자

```python
X1 = data[['gender', 'region']]  # 인코딩할 변수만 따로 분리
XY = data[['edu', 'income', 'age', 'score_gov', 'score_progress', 'score_intention', 'vote', 'parties']]
```

### 원핫인코딩

```python
X1.loc[:, 'gender'] = X1.loc[:, 'gender'].replace([1,2], ['male', 'female'])
X1.loc[:, 'region'] = X1.loc[:, 'region'].replace([1,2,3,4,5], ['Sudo', 'Chungcheung', 'Honam', 'Youngnam', 'Others'])
X1.head()
```

```python
X1_dum = pd.get_dummies(X1)
X1_dum.head()
```

```python
X2 = data
X2.loc[:, 'gender'] = X2['gender'].replace([1,2], ['male', 'female'])
X2.loc[:, 'region'] = X2['region'].replace([1,2,3,4,5], ['Sudo', 'Chungcheung', 'Honam', 'Youngnam', 'Others'])

X2_dum = pd.get_dummies(X2, columns=['gender', 'region'])  # columns로 명식적으로 지정할 수도 있음
X2_dum.head()
```

### 자료 저장

```python
X2.to_csv('../data/Fvote.csv', index=False, sep=',')
```
